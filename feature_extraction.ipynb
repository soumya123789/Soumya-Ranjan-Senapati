# importing all required libraries

```python
import librosa
import librosa.display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
```

```python
mylist = glob('audio_data/'+ '/*.wav')   # import ing all audio data into a list called "mylist"
```

```python
len(mylist)
```

# Plotting a single audio file 

```python
data, sampling_rate = librosa.load('audio_data/a21.wav')
plt.figure(figsize=(15, 5))
librosa.display.waveplot(data, sr=sampling_rate)
```

# Data extraction from all audio files

```python
a_list = []

for i in range(len(mylist)):
    y_i , sr_i = librosa.load(mylist[i], duration=2, sr=44100)
    a_list.append(y_i)
```

```python
a_list
```

# Converting a_list into dataframe and nd

```python
audio_dataframe = pd.DataFrame(a_list)
audio_array = np.array(a_list)
```

```python
audio_dataframe
```

## Extracting spectral features (mfcc, mfcc delta, mfcc delta square) of each audio files

```python
mfccs = []
mfcc_delta_1 = []
mfcc_delta_2 = []
for i in range(len(audio_array)):
    feature_i = librosa.feature.mfcc(y=audio_array[i], sr= 44100, n_mfcc=13)
    mfcc_i = np.mean(feature_i)
    mfccs.append(mfcc_i)
    delta_1 = librosa.feature.delta(feature_i, order=1)
    delta_1_mean = np.mean(delta_1)
    mfcc_delta_1.append(delta_1_mean)
    delta_2 = librosa.feature.delta(feature_i, order=2)
    delta_2_mean = np.mean(delta_2)
    mfcc_delta_2.append(delta_2_mean)
```

```python
feature_mfcc = pd.DataFrame(mfccs, columns=['mfcc'])
feature_delta_1 = pd.DataFrame(mfcc_delta_1, columns=['delta_1'])
feature_delta_2 = pd.DataFrame(mfcc_delta_2, columns=['delta_2'])
```

```python
audio_features = pd.concat([feature_mfcc, feature_delta_1, feature_delta_2], axis=1)
```

```python
df = pd.read_csv('label_data.csv')
audio_labels = df.label
```

```python
speech_data = pd.concat([audio_features, audio_labels], axis=1)
```

```python
speech_data.to_csv('speech_emotion_recognition.csv', index=False)
```
